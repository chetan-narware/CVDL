```python
import pandas as pd
import cv2 as cv
import numpy as np
```

```python
print(cv.__version__)
```

```python
img = cv.imread("simg.jpg")
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.imshow(img)
```

```python
cv.imshow("cat_image",img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img2 = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
img2.shape
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)
ax.imshow(img2)
ax.axis('off')
ax.set_title("Image Display")
plt.show()
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
plt.imshow(gray, cmap='gray')
plt.show()
```

```python
plt.imshow(img2[:,:,1],cmap='gray')
plt.show()
```

```python
resized = cv.resize(img2,(300,300))
plt.imshow(resized)
```

```python
line = cv.line(img2,(25,25),(1175,1175),(255,0,0),thickness = 10)
plt.imshow(line)
```

```python
cv.line(resized, (50, 50), (200, 50), (255, 0, 0), 5)
cv.rectangle(resized, (60, 60), (160, 160), (0, 255, 0), 3)
cv.circle(resized, (120, 120), 40, (0, 0, 255), -1)
plt.imshow(resized)
plt.show()
```
